```python
with open("img_celeba_data/list_bbox_celeba.txt") as f:
    bbox_data = f.readlines()
```

```python
bbox_data = bbox_data[2:]
```

```python
bbox_points = dict()
for point in bbox_data:
    img_path, x,y, w, h = point.split()
    
    bbox_points[img_path] = (x,y,w,h)
```

```python
with open("img_celeba_data/list_identity_celeba.txt") as f:
    label_data = f.readlines()
```

```python
label_data = label_data[2:]
```

```python
label_points = dict()
for point in label_data:
    img_path, label = point.split()
    
    if label not in label_points:
        label_points[label] = [img_path]
    else:
        label_points[label].append(img_path)
```

```python
mapping = dict()

i = 1
for label in label_points:
    length = len(label_points[label])
    if length >= 30:
        mapping[label] = i
        i+=1
    else:
        mapping[label] = 0
```

```python
import json
    
with open("img_celeba_data/mapping.json", "w") as outfile:
    json.dump(mapping, outfile)
```

```python
label_to_imgs = dict()

for point in label_data:
    img_path, label = point.split()
    
    int_label = mapping[label]
    
    if int_label not in label_to_imgs:
        label_to_imgs[int_label] = [img_path]
    else:
        label_to_imgs[int_label].append(img_path)
```

```python
label_to_imgs
```

```python
for label,img_path in label_to_imgs.items():
    import random
    random.shuffle(img_path)
    
    test_paths = img_path[:5]
    train_paths = img_path[5:100]
    
    for test_path in test_paths:
        save_dict = dict()
        save_dict["label"] = label
        save_dict["box"] = bbox_points[test_path]
        save_dict["path"] = test_path
        with open("img_celeba_data/val/"+ test_path[:-4] + ".json", "w") as outfile:
            json.dump(save_dict, outfile)
        
    for train_path in train_paths:
        save_dict = dict()
        save_dict["label"] = label
        save_dict["box"] = bbox_points[train_path]
        save_dict["path"] = train_path
        with open("img_celeba_data/train/"+ train_path[:-4] + ".json", "w") as outfile:
            json.dump(save_dict, outfile)
```
